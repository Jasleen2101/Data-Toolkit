## Part A — Conceptual Questions

### Q1. What is NumPy, and why is it widely used in Python?

**Answer:** **NumPy** (Numerical Python) is a core scientific computing library that provides:
- Fast **N-dimensional arrays** (`ndarray`)
- Vectorized math operations
- Linear algebra, random numbers, Fourier transforms

It’s widely used because it’s **much faster than Python loops** for numeric work and forms the base for Pandas, SciPy, scikit-learn, etc.

### Q2. How does broadcasting work in NumPy?

**Answer:** Broadcasting is NumPy’s rule system that allows operations between arrays of different shapes by **virtually expanding** one array to match the other (without copying data).  
Two dimensions are compatible if they are equal or one of them is 1. Example: `(3,1)` + `(1,4)` → `(3,4)`.

### Q3. What is a Pandas DataFrame?

**Answer:** A **Pandas DataFrame** is a 2D labeled tabular data structure (rows + columns) where:
- Columns can be different data types
- It supports indexing, filtering, grouping, joining, reshaping, and time-series tools.

### Q4. Explain the use of the groupby() method in Pandas.

**Answer:** `groupby()` splits data into groups based on one or more keys and then applies an aggregation/transform.  
Example uses: sum/mean per category, count per group, custom aggregations, multi-level grouping.

### Q5. Why is Seaborn preferred for statistical visualizations?

**Answer:** Seaborn is preferred for statistical plots because it:
- Works smoothly with Pandas DataFrames
- Has built-in statistical aesthetics (confidence intervals, regression, distributions)
- Produces good-looking plots with less code.

### Q6. What are the differences between NumPy arrays and Python lists?

**Answer:** **NumPy arrays**: homogeneous, fixed-size, vectorized ops, memory-efficient, fast slicing.  
**Python lists**: heterogeneous, dynamic size, slower numeric operations (loop required).

### Q7. What is a heatmap, and when should it be used?

**Answer:** A **heatmap** displays values as colors in a matrix layout (e.g., correlations, confusion matrices).  
Use it when you want to quickly spot patterns, high/low regions, or relationships between variables.

### Q8. What does the term “vectorized operation” mean in NumPy?

**Answer:** A **vectorized operation** performs computations on entire arrays at once (implemented in optimized C) instead of looping element-by-element in Python.

### Q9. How does Matplotlib differ from Plotly?

**Answer:** Matplotlib is a low-level, highly customizable static plotting library.  
Plotly is interactive (zoom, hover tooltips), web-friendly, and great for dashboards (but may be heavier).

### Q10. What is the significance of hierarchical indexing in Pandas?

**Answer:** Hierarchical (MultiIndex) indexing allows multiple index levels in rows/columns, enabling:
- More complex grouping/reshaping
- Pivot-style analysis
- Efficient slicing on multi-dimensional keys.

### Q11. What is the role of Seaborn’s pairplot() function?

**Answer:** `pairplot()` creates a grid of scatterplots for pairwise relationships and hist/KDE on diagonals—useful for quick EDA and spotting correlations/outliers.

### Q12. What is the purpose of the describe() function in Pandas?

**Answer:** `describe()` gives summary stats like count, mean, std, min, quartiles, max for numeric columns (and frequency summaries for object columns).

### Q13. Why is handling missing data important in Pandas?

**Answer:** Missing data affects calculations and models (mean, correlations, ML training). Handling it prevents bias/errors and improves data quality (fill, drop, interpolate).

### Q14. What are the benefits of using Plotly for data visualization?

**Answer:** Plotly provides:
- Interactive visuals (hover, zoom, pan)
- Easy sharing in HTML
- Advanced chart types (3D, animated, maps)
- Useful for dashboards and presentations.

### Q15. How does NumPy handle multidimensional arrays?

**Answer:** NumPy supports multidimensional arrays (`ndarray`) with:
- Shape, dtype, strides for memory layout
- Fast slicing/views
- Broadcasting and vectorized ops across dimensions.

### Q16. What is the role of Bokeh in data visualization?

**Answer:** Bokeh is used for interactive, browser-based visualizations with Python—good for dashboards and streaming/real-time plots.

### Q17. Explain the difference between apply() and map() in Pandas.

**Answer:** `map()` is mainly for **Series element-wise** mapping using a dict/function.  
`apply()` can work row/column-wise on DataFrames (more flexible, can apply complex functions).

### Q18. What are some advanced features of NumPy?

**Answer:** Advanced NumPy features include:
- Broadcasting, fancy indexing, boolean masking
- `einsum`, ufuncs, vectorization
- Memory views, strides tricks
- Linear algebra & FFT modules.

### Q19. How does Pandas simplify time series analysis?

**Answer:** Pandas simplifies time series with:
- DateTimeIndex, resampling (`resample`), rolling windows (`rolling`)
- Time-aware slicing (e.g., `df['2024-01':'2024-03']`)
- Time zone handling.

### Q20. What is the role of a pivot table in Pandas?

**Answer:** Pivot tables summarize data by categories using aggregations—quick way to create Excel-like summaries inside Pandas (`pivot_table`).

### Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

**Answer:** NumPy slicing is fast because arrays are contiguous blocks of memory and slicing often creates **views** (no copy) + operations run in optimized C.

### Q22. What are some common use cases for Seaborn?

**Answer:** Common Seaborn use cases: distributions (hist/KDE), categorical comparisons (box/violin), regression plots, correlation heatmaps, pairplots for EDA.

### Practical Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python

df_corr = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 2, 3, 2],
})

corr = df_corr.corr(numeric_only=True)
plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

corr
```
